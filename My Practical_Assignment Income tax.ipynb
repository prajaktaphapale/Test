### How income tax is calculated1. 
From total annual income a standard deduction of 50000 INR is don
2. Apply the tax according to slabs
3. apply 4% cess on the tax
4. total tax = tax + cesse applied

### Assumptions

1. Person is salaried indiviual.
2. Age is less than 60
3. No Tax exemptions applied

```python
# Take total income as input from user

income = float(input("Please provide the annual income in INR : "))

# Apply standard deduction of 50000
std_ded = 50000
ti = income - std_ded

# Print total income and taxable income
print (f"Total Income : {income:.2f} INR, Taxable income : {ti:.2f}")

# Apply if else loop based on taxable income only

if ti<=700000:
    tax=0
    print(f"According to new Regime Tax is not applicable up to taxable income 7 lac")

elif ti>300000 and ti<=600000:
    tax = (5/100)*(ti - 300000)
    print(f'For taxable income of {ti:.2f} INR the tax slab is 5%')
elif ti>600000 and ti<=900000:
    tax = 15000 + (10/100)*(ti-600000)
    print(f'For taxable income of {ti:.2f} INR the tax slab is 10%')
elif ti>900000 and ti<=1200000:
    tax = 15000 + 30000 + (15/100)*(ti-900000)
    print(f'For taxable income of {ti:.2f} INR the tax slab is 15%')
elif ti>1200000 and ti<=1500000:
    tax = 15000 + 30000 + 45000 + (20/100)*(ti-1200000)
    print(f'For taxable income of {ti:.2f} INR the tax slab is 20%')
else:
    tax = 15000 + 30000 + 45000 + 60000 + (30/100)*(ti-1500000)
    print(f'For taxable income of {ti:.2f} INR the tax slab is 30%')

# Add 4 % of cess
cess = (4/100)*tax

# Total tax
total_tax = tax + cess

# Print all the results
print(f'Tax Before cess : {tax:.2f} INR')
print(f'Cess 4% : {cess:.2f} INR')
print(f'Total Tax : {total_tax:.2f} INR')

```
